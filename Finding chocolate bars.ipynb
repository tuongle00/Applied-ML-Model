## Data science notebooks & data analysis
Notebooks are great for interactive data analysis. Let's create a pandas DataFrame using the `read_csv()` function. 

We will load the dataset "chocolate_bars.csv" containing different characteristics of over 2,500 chocolate bars and their reviews.

By using the `.head()` method, we display the first five rows of data:

```python
# Importing the pandas module
import pandas as pd

# Reading in the data
df = pd.read_csv('data/chocolate_bars.csv')

# Take a look at the first datapoints
df.head()
```

## Data analysis example:

Find the average rating for chocolate bars with different numbers of ingredients. 

We can use `.groupby()` to group the information by the column "num_ingredients". Then we select the column "rating" and use `.mean()` to get the average rating for each group:

```python
df.groupby('num_ingredients')[['rating']].mean()
```

## Data science notebooks & visualizations
Visualizations are very helpful to summarize data and gain insights. A well-crafted chart often conveys information much better than a table.

It is very straightforward to include plots in a data science notebook. For example, let's take a look at the relationship between review year and rating. 

We are using the `seaborn` library for this example. We will run the `scatterplot()` function and include the variables we want to display.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.scatterplot(x='year_reviewed', y='rating', data=df)
plt.show()
```

We can also make a plot from the table we calculated above (average rating for chocolate bars with different numbers of ingredients):

```python
sns.barplot(x='num_ingredients', y='rating', data=df)
plt.show()
```

```python

#Challenge 1:What is the average rating by country of origin?
import numpy as np

#Find Unique Origin
unique_origin = pd.unique(df['bean_origin'])

sort = np.sort(unique_origin)
print(sort)

print('')

#Define to get count based on origin
def get_avg_rating(origin):
    origin_rating = df.loc[(df['bean_origin'] == origin)]
    return origin_rating.mean()['rating']


#Count rating average by input the origin
for origin in sort:
    print(origin,":",get_avg_rating(origin))



    
```

Challenge 1 show that bean origin does have an effect on bean quality

```python
#Challenge 2:How many bars were reviewed for each of those countries?

def get_bar_name(origin):
    origin_rating = df.loc[(df['bean_origin'] == origin)]
    return origin_rating.count()['bar_name']


for origin in sort:
    print(origin,":",get_bar_name(origin))


```

However Challenge 2 shows that the data is bias toward some countries bean origin where some countries has more bar name review than other which could pump up the rating without having enough sample size to determine the rating

```python
#Challenge 3: Create plots to visualize findings for questions 1 and 2.
df_group = df.groupby('bean_origin')
bar_name_count = df_group.count()['bar_name']
rating_count = df_group.mean()['rating']
groups = [group for group,df in df_group]


#Plot Challenge 1
plt.figure(figsize=(15,8))
ax =sns.barplot(x = groups, y = rating_count)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90 )
```

```python
#Plot Challenge 2

plt.figure(figsize=(15,8))
ax =sns.barplot(x = groups, y = bar_name_count)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90 )
```

#Challenge 4:Is the cacao bean's origin an indicator of quality?

No because sample size need to be bigger to determine the quality.

```python
#Challenge 5:[Optional] How does cocoa content relate to rating?
#What is the average cocoa content for bars with higher ratings (above 3.5)?

have_cocoa = df.loc[df['review'].str.contains('cocoa')]
no_cocoa = df.loc[~df['review'].str.contains('cocoa')]


avg_rating_have_cocoa = have_cocoa.mean()['rating']
avg_rating_no_cocoa = no_cocoa.mean()['rating']

print("Average rating of content with cocoa:",avg_rating_have_cocoa)
print("Average rating of content without cocoa :",avg_rating_no_cocoa)


have_cocoa_high_rating_avg = df.loc[(df['review'].str.contains('cocoa')) &(df['rating']> 3.5)].mean()['rating']

print("Average rating cocoa content for bars with high rating 3.5 or above:",have_cocoa_high_rating_avg)
```

```python
#Challenge 6:[Optional 2] Your research indicates that some consumers want to avoid bars with lecithin. Compare the average rating of bars with and without lecithin (L in the ingredients).

have_L = df.loc[df['ingredients'].str.contains('L',na = False)]
no_L = df.loc[~df['ingredients'].str.contains('L',na = False)]

have_L_avg_rating = have_L.mean()['rating']
no_L_avg_rating = no_L.mean()['rating']

print("Average rating for bars with lecithin:",have_L_avg_rating )
print("Average rating for bars without lecithin:",no_L_avg_rating)
```
